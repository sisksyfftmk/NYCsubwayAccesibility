Are New Yorkers Strong? The Reality of Subway Accessibility

### Data Dictionary & Methodology: Defining "Accessibility"

In the MTA dataset, the `ADA` column indicates the accessibility status of each station based on the Americans with Disabilities Act (ADA) standards. The values are defined as follows:

* **0:** Not ADA-accessible (No elevators/ramps; not accessible for wheelchair users)
* **1:** Fully accessible (Equipped with elevators/ramps; fully accessible to all platforms)
* **2:** Partially accessible (Accessibility is limited to specific lines, directions, or entrances/exits)

**Why I only count "1" (Fully Accessible) in this analysis:**

While "2" (Partially accessible) exists in the dataset, it represents a significant, unpredictable barrier for those relying on wheelchairs or strollers. Partial accessibility often means users might arrive at a station only to find they cannot reach their specific train, transfer to another line, or use their desired exit without extensive prior research. In the context of urban mobility, a "partially accessible" station is fundamentally unreliable. Therefore, to truly measure equitable infrastructure, this project strictly defines an accessible station as one with an `ADA` value of `1`.

```python
borough_stats = df.pivot_table(index='Borough',columns='ADA',aggfunc='size',fill_value=0)

borough_stats['Total'] = borough_stats.sum(axis=1)
borough_stats['Accessibility_Rate']=(borough_stats.get(1))/borough_stats['Total']*100
print(borough_stats[['Total',1,'Accessibility_Rate']])
```

```python
params3 = {
    "get": "NAME,DP05_0029E",
    "for": "county:005,047,061,081,085",
    "in": "state:36",
    "key": api_key
}

response3 = requests.get(url, params=params3)

data3 = response3.json()
senior_pop_df = pd.DataFrame(data3[1:], columns=data3[0])

senior_pop_df = senior_pop_df[['NAME', 'DP05_0029E']]
senior_pop_df['DP05_0029E'] = senior_pop_df['DP05_0029E'].astype(int)
senior_pop_df = senior_pop_df.rename(columns={'DP05_0029E': 'Pop_65_and_Over'})

senior_pop_df
```

```python
merged_df = pd.merge(total_pop_df, under_5_df, on="NAME")
merged_df = pd.merge(merged_df, senior_pop_df, on="NAME")

merged_df["Pct_Under_5"] = (merged_df["Pop_Under_5"] / merged_df["Total_Population"]) * 100
merged_df["Pct_65_and_Over"] = (merged_df["Pop_65_and_Over"] / merged_df["Total_Population"]) * 100

merged_df["Pop_Vulnerable_Total"] = merged_df["Pop_Under_5"] + merged_df["Pop_65_and_Over"]
merged_df["Pct_Vulnerable_Total"] = merged_df["Pct_Under_5"] + merged_df["Pct_65_and_Over"]

demographics_df = merged_df[[
    "NAME", 
    "Total_Population", 
    "Pop_Under_5", 
    "Pct_Under_5", 
    "Pop_65_and_Over", 
    "Pct_65_and_Over",
    "Pop_Vulnerable_Total",   
    "Pct_Vulnerable_Total"    
]]

demographics_df
```
